### Imports

```python
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon
from highlight_text.htext import htext
from math import pi
```

### Logo

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal') # set this before using htext
plt.ylim(0., 1)
plt.xlim(0., 1)

poly = RegularPolygon(xy = (0.5, 0.5), numVertices = 6, radius = 0.45, orientation = pi,
                      edgecolor = 'k', facecolor = '#f2f2f2', lw = 5)
ax.add_patch(poly)

texts = htext(s = '<HIGHLIGHT>\nTEXT', x = 0.5, y = 0.5, color = 'k', fontsize = 24,
      highlight_colors = ['#ff8600'], highlight_weights = ['bold'],
      highlight_styles = ['italic'], string_weight = 'regular',
      ha = 'center', va = 'center')

plt.axis('off');
```
